```python
import numpy as np
import pandas as pd
from time import time
from IPython.display import display
import matplotlib.pyplot as plt
# import seaborn as sns
# import visuals as vs
```

```python
# Load the Red Wines dataset
data = pd.read_csv("static/data/clean_winequality.csv")

data.head()
```

```python
data.dtypes
```

```python
#need

n_wines = data.shape[0]

# Number of wines with quality rating above 7
quality_above_7 = data.loc[(data['quality'] >= 7)]
n_above_7 = quality_above_7.shape[0]

# Number of wines with quality rating below 3
quality_below_3 = data.loc[(data['quality'] <= 3)]
n_below_3 = quality_below_3.shape[0]

# Number of wines with quality rating between 4 to 6
quality_between_4 = data.loc[(data['quality'] >= 4) & (data['quality'] <= 6)]
n_between_4 = quality_between_4.shape[0]

# Percentage of wines with quality rating above 7
greater_percent = n_above_7*100/n_wines

# Print the results
print("Total number of wine data: {}".format(n_wines))
print("Wines with rating 6 and above: {}".format(n_above_7))
print("Wines with rating less than 3: {}".format(n_below_3))
print("Wines with rating 4 and 6: {}".format(n_between_4))
print("Percentage of wines with quality 7 and above: {:.2f}%".format(greater_percent))
```

```python
from scipy.stats import linregress
import scipy.stats as st
```

```python
volatile_acidity_quality = red[['volatile acidity','quality']]
fig, axs = plt.subplots(ncols=1,figsize=(10,6))
sns.barplot(x='quality', y='volatile acidity', data=volatile_acidity_quality, ax=axs)
plt.title('volatile acidity VS quality')

plt.tight_layout()
plt.show()
plt.gcf().clear()

print("A higher quality is usually associated with low volatile acidity levels because volatile acidity is an indicator of spoilage and could give rise to unpleasant aromas — consistent with our domain knowledge.")
```

```python
citric_acid_quality = red[['citric acid','quality']]
fig, axs = plt.subplots(ncols=1,figsize=(10,6))
sns.barplot(x='quality', y='citric acid', data=citric_acid_quality, ax=axs)
plt.title('citric acid VS quality')

plt.tight_layout()
plt.show()
plt.gcf().clear()

print("higher citric acid has better quality")
```

```python
chlorides_quality = red[['chlorides','quality']]
fig, axs = plt.subplots(ncols=1,figsize=(10,6))
sns.barplot(x='quality', y='chlorides', data=chlorides_quality, ax=axs)
plt.title('chlorides VS quality')

plt.tight_layout()
plt.show()
plt.gcf().clear()

print("seems lower chlorides has better quality")
```

```python
sulfates_quality = red[['sulfates','quality']]
fig, axs = plt.subplots(ncols=1,figsize=(10,6))
sns.barplot(x='quality', y='sulfates', data=sulfates_quality , ax=axs)
plt.title('chlorides VS quality')

plt.tight_layout()
plt.show()
plt.gcf().clear()

print("seems higher chlorides has better quality but not a deal breaker")
```

```python
alcohol_quality = red[['alcohol','quality']]
fig, axs = plt.subplots(ncols=1,figsize=(10,6))
sns.barplot(x='quality', y='alcohol', data=alcohol_quality, ax=axs)
plt.title('alcohol VS quality')

plt.tight_layout()
plt.show()
plt.gcf().clear()

print("Seems like most people generally like wines that contain a higher percentage of alcohol")
```

Visualize the correlation  - components that have correlation

```python
fixed_acidity_citric_acid = red[['fixed acidity','citric acid']]

gridA = sns.JointGrid(x="fixed acidity", y="citric acid", data=fixed_acidity_citric_acid, size=6)

gridA = gridA.plot_joint(sns.regplot, scatter_kws={"s": 10})

gridA = gridA.plot_marginals(sns.distplot)


xvalue = red["fixed acidity"]
yvalue = red["citric acid"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(xvalue, yvalue)
regress = xvalue*slope + intercept

line = "y = " + str(round(slope,2)) + "x+" + str(round(intercept,2))

print(f'The r-value is : {rvalue**2}')

correlation = st.pearsonr(xvalue, yvalue) 
print(f'correlation is {round(correlation[0],2)}')

#Strenght of Correlation:   r<0.3 - none/very weak     0.3<= r <0.5 - weake    0.5<= r <0.7 - moderate    r>=0.7 - strong

print("there is moderate correlation between fixed acidity and citric acid")
```
